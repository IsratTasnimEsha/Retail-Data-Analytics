```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.chdir('Resources/')
```

```python
df_r2 = pd.read_csv("2_9_Client_r2_History.csv")

line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-']
marker_styles = ['o', 's', 'D', '^', 'v', '<', '>', 'P', '*']
gray_shades = ['0.0', '0.07', '0.14', '0.21', '0.24', '0.35', '0.42', '0.49', '0.56']

plt.figure(figsize=(14, 7))
sns.set(style="whitegrid")

for i, client in enumerate(sorted(df_r2["Client"].unique())):
    client_data = df_r2[df_r2["Client"] == client]
    plt.plot(
        client_data["Round"],
        client_data["R² Score"],
        label=client,
        linestyle=line_styles[i % len(line_styles)],
        marker=marker_styles[i % len(marker_styles)],
        color=gray_shades[i % len(gray_shades)],
        linewidth=1.5
    )

plt.title("R² Score per Client Over Federated Rounds", fontsize=20)
plt.xlabel("Round", fontsize=15)
plt.ylabel("R² Score", fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(
    title="Clients",
    loc='center left',
    bbox_to_anchor=(1.0, 0.5),
    fontsize=11,
    title_fontsize=13,
    frameon=True
)

plt.tight_layout()
plt.savefig("Images/3_r2_Score_9_Clients_Encoded.jpg", format="jpg", dpi=300)
plt.show()
```
