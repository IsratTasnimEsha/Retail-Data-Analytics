```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df_r2 = pd.read_csv("2_15_Client_r2_History.csv")

line_styles = ['-', '--', '-.', ':'] * 4
marker_styles = ['o', 's', 'D', '^', 'v', '<', '>', 'P', '*', 'X', 'h', 'H', 'x', '+', '|']
gray_shades = [
    '0.0', '0.5', '0.1', '0.15', '0.2',
    '0.25', '0.3', '0.35', '0.4', '0.45',
    '0.5', '0.55', '0.6', '0.65', '0.7'
]

plt.figure(figsize=(14, 7))
sns.set(style="whitegrid")

clients_sorted = sorted(df_r2["Client"].unique(), key=lambda x: int(x.split()[1]))

for i, client in enumerate(clients_sorted):
    client_data = df_r2[df_r2["Client"] == client]
    plt.plot(
        client_data["Round"],
        client_data["R Score"],
        label=client,
        linestyle=line_styles[i % len(line_styles)],
        marker=marker_styles[i % len(marker_styles)],
        color=gray_shades[i % len(gray_shades)],
        linewidth=1.5
    )

plt.title("R Score per Client Over Federated Rounds", fontsize=20)
plt.xlabel("Round", fontsize=15)
plt.ylabel("R Score", fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(
    title="Clients",
    loc='center left',
    bbox_to_anchor=(1.0, 0.5),
    fontsize=11,
    title_fontsize=13,
    frameon=True
)

plt.tight_layout()
plt.savefig("Images/3_r2_Score_15_Clients.jpg", format="jpg", dpi=300)
plt.show()
```
